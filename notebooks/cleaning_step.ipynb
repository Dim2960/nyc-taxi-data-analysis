# Nettoyage des donn√©es


```python
import pandas as pd
import datetime
```

```python
file_path = "../data_clean/taxi_trip_2016-preclean.parquet"

df = pd.read_parquet(file_path)
```

```python
df['id'] = df.index + 1
```

```python
df.head(10)
```

```python
df['trip_time'] = df['tpep_dropoff_datetime'] - df['tpep_pickup_datetime']
df['trip_time'] = df['trip_time'].dt.total_seconds() 
```

```python
df['tpep_pickup_datetime'] = df['tpep_pickup_datetime'].dt.floor('h')
df['tpep_dropoff_datetime'] = df['tpep_dropoff_datetime'].dt.floor('h')
```

```python
df['pickup_hour'] = df['tpep_pickup_datetime'].dt.hour
```

```python
df['date'] = df['tpep_pickup_datetime'].dt.date
```

```python
df.head(10)
```

```python
df.info()
```

## 2 - cr√©ation des table de dimension borough et neighborhood

```python
df_dim_borough = pd.DataFrame(df['pickup_borough'].unique(), columns=['borough'])
df_dim_borough['id_borough'] = df_dim_borough.index + 1
```

```python
df_dim_borough.fillna('Autres', inplace=True)
```

```python
df_dim_borough
```

```python
# Extraire les valeurs uniques des quartiers et des boroughs
df_dim_neighborhood = df[['pickup_neighborhood', 'pickup_borough']].drop_duplicates()

# Renommer les colonnes
df_dim_neighborhood.columns = ['neighborhood', 'borough']

df_dim_neighborhood['id_neighborhood'] = df_dim_neighborhood.index + 1

df_dim_neighborhood = df_dim_neighborhood[['id_neighborhood', 'neighborhood', 'borough']]
```

```python
df_dim_neighborhood = df_dim_neighborhood.merge(df_dim_borough, on='borough', how='left') 

df_dim_neighborhood.drop(columns=['borough'], axis=1, inplace=True)
```

```python
df_dim_neighborhood['neighborhood'].fillna('Autres', inplace=True)
```

```python
df_dim_neighborhood
```

## 3 - Normalisation des borouh et neighborhood

```python
df = df.merge(df_dim_borough, left_on='pickup_borough', right_on='borough', how='left') 
df.rename(columns={'id_borough': 'pickup_borough_id'}, inplace=True)
df.drop(columns=['pickup_borough', 'borough'], axis=1, inplace=True)
```

```python
df = df.merge(df_dim_neighborhood, left_on='pickup_neighborhood', right_on='neighborhood', how='left')
df.rename(columns={'id_neighborhood': 'pickup_neighborhood_id'}, inplace=True)
df.drop(columns=['pickup_neighborhood', 'neighborhood'], axis=1, inplace=True)
```

```python
df = df.merge(df_dim_borough, left_on='dropoff_borough', right_on='borough', how='left')
df.rename(columns={'id_borough': 'dropoff_borough_id'}, inplace=True)
df.drop(columns=['dropoff_borough', 'borough'], axis=1, inplace=True)
```

```python
df = df.merge(df_dim_neighborhood, left_on='dropoff_neighborhood', right_on='neighborhood', how='left')
df.rename(columns={'id_neighborhood': 'dropoff_neighborhood_id'}, inplace=True)
df.drop(columns=['dropoff_neighborhood', 'neighborhood'], axis=1, inplace=True)
```

```python
df.head(6)
```

## 4 - Generation table de fait

```python
df_fact = df[[
        'id',
        'VendorID', 
        'date',
        'tpep_pickup_datetime', 
        'pickup_hour',
        'passenger_count', 
        'trip_distance', 
        'trip_time',
        'RatecodeID', 
        'store_and_fwd_flag',
        'payment_type',  
        'pickup_neighborhood_id', 
        'dropoff_neighborhood_id',
        'fare_amount',
        'extra',
        'mta_tax',
        'tip_amount',
        'tolls_amount',
        'improvement_surcharge',
        'total_amount'
    ]].copy()
```

```python
df_fact.head()
```

## 5 - Enregistrement parquet


```python
df_fact.to_parquet("../data_clean/taxi_trip_2016-clean1.parquet", index=False)
```
